```python
from openai import OpenAI
import pandas as pd
import pyodbc

# BANCO DE DADOS ----------------------------------------------------------------
# Armazena variáveis do Banco de Dados salvos em excel
df_config = pd.read_excel("config.xlsx") # PoderIA ser criptografado
driver = df_config.loc[df_config['nome'] == 'DRIVER', 'Valor'].values[0]
server = df_config.loc[df_config['nome'] == 'SERVER', 'Valor'].values[0]
database = df_config.loc[df_config['nome'] == 'DATABASE', 'Valor'].values[0]
tabela = df_config.loc[df_config['nome'] == 'TABLE', 'Valor'].values[0]

del df_config #Limpa memória do sistema

# abre conexão DB
conn = pyodbc.connect(
    f"DRIVER={driver};"
    f"SERVER={server};"
    f"DATABASE={database};"
    "Trusted_Connection=yes;"
)

query = f"SELECT * FROM {tabela}"
df = pd.read_sql(query, conn)
conn.close() # fecha conexão com DB

#display(df) #Caso queira verificar os dados
# ------------------------------------------------------------------------------
```
